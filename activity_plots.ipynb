```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
daily_follow_config_1 = pd.read_csv('data/exp_set_1/follows_daily_sum.csv')
daily_follow_config_2 = pd.read_csv('data/exp_set_2/follows_daily_sum.csv')
```

```python
fig, ax = plt.subplots(figsize=(8, 5))

ax.plot(
    daily_follow_config_1['day'],
    daily_follow_config_1['n_follows'],
    label='Config 1'
)

ax.plot(
    daily_follow_config_2['day'],
    daily_follow_config_2['n_follows'],
    label='Config 2'
)

ax.set_title('Daily follows')
ax.set_xlabel('Day')
ax.set_ylabel('Number of follows')
ax.legend()

plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8, 5))

ax.plot(
    daily_follow_config_1['day'],
    daily_follow_config_1['n_unfollows'],
    label='Config 1'
)

ax.plot(
    daily_follow_config_2['day'],
    daily_follow_config_2['n_unfollows'],
    label='Config 2'
)

ax.set_title('Daily unfollows')
ax.set_xlabel('Day')
ax.set_ylabel('Number of unfollows')
ax.legend()

plt.tight_layout()
plt.show()
```

```python
daily_follow_1 = list(daily_follow_config_1['n_follows'])
daily_follow_2 = list(daily_follow_config_2['n_follows'])

daily_follow_1 = [daily_follow_1[0]] + [
    daily_follow_1[i] - daily_follow_1[i-1]
    for i in range(1, len(daily_follow_1))
]

daily_follow_2 = [daily_follow_2[0]] + [
    daily_follow_2[i] - daily_follow_2[i-1]
    for i in range(1, len(daily_follow_2))
]

days_1 = np.arange(1, len(daily_follow_1) + 1)
days_2 = np.arange(1, len(daily_follow_2) + 1)

plt.figure(figsize=(10, 4))

plt.plot(days_1, daily_follow_1, label='Config 1')
plt.plot(days_2, daily_follow_2, label='Config 2')

plt.xlabel('Day')
plt.ylabel('Number of follow')
plt.title('Follows over time')
plt.legend()
plt.grid(alpha=0.3)

plt.tight_layout()
plt.show()


daily_unfollow_1 = list(daily_follow_config_1['n_unfollows'])
daily_unfollow_2 = list(daily_follow_config_2['n_unfollows'])

daily_unfollow_1 = [daily_unfollow_1[0]] + [
    daily_unfollow_1[i] - daily_unfollow_1[i-1]
    for i in range(1, len(daily_unfollow_1))
]

daily_unfollow_2 = [daily_unfollow_2[0]] + [
    daily_unfollow_2[i] - daily_unfollow_2[i-1]
    for i in range(1, len(daily_unfollow_2))
]

days_1 = np.arange(1, len(daily_unfollow_1) + 1)
days_2 = np.arange(1, len(daily_unfollow_2) + 1)

plt.figure(figsize=(10, 4))

plt.plot(days_1, daily_unfollow_1, label='Config 1')
plt.plot(days_2, daily_unfollow_2, label='Config 2')

plt.xlabel('Day')
plt.ylabel('Number of unfollow')
plt.title('Unfollows over time')
plt.legend()
plt.grid(alpha=0.3)

plt.tight_layout()
plt.show()
```
